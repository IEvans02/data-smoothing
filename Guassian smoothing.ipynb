```python
import numpy as np
from scipy.ndimage import gaussian_filter
```

```python
def gaussian_smoothing(image, sigma):
    """
    Applies Gaussian smoothing to the given image using a Gaussian filter with the specified sigma.
    
    Args:
        image (numpy array): Input image to be smoothed.
        sigma (float): Standard deviation of the Gaussian filter.
        
    Returns:
        smoothed_image (numpy array): Smoothed image.
    """
    smoothed_image = gaussian_filter(image, sigma=sigma)
    return smoothed_image
```

```python
# Load an example image
from skimage import data
image = data.camera()

# Apply Gaussian smoothing with sigma=2
smoothed_image = gaussian_smoothing(image, sigma=3)

# Display the original and smoothed images side by side
import matplotlib.pyplot as plt
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(8, 4))
ax1.imshow(image, cmap='gray')
ax1.set_title('Original')
ax2.imshow(smoothed_image, cmap='gray')
ax2.set_title('Smoothed')
plt.show()
```
