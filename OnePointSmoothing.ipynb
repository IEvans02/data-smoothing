```python
import math

def point_boundary_smoothing(points, angle_threshold, distance_threshold):
    new_points = []
    n = len(points)
    i = 0
    while i < n:
        new_points.append(points[i])
        if i < n - 2:
            x1, y1 = points[i]
            x2, y2 = points[i+1]
            x3, y3 = points[i+2]
            if (x2 - x1) * (y3 - y2) == (y2 - y1) * (x3 - x2):
                # Points are collinear
                i += 1
            else:
                angle = math.degrees(math.atan2(y3 - y2, x3 - x2) - math.atan2(y1 - y2, x1 - x2))
                if abs(angle) < angle_threshold:
                    d1 = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
                    d2 = math.sqrt((x3 - x2)**2 + (y3 - y2)**2)
                    if d1 < distance_threshold or d2 < distance_threshold:
                        i += 1
        i += 1
    return new_points
```

```python
points = [(0, 0), (1, 1), (2, 2), (3, 3), (4, 4)]
angle_threshold = 10 # degrees
distance_threshold = 0.5 # units
smoothed_points = point_boundary_smoothing(points, angle_threshold, distance_threshold)
print(smoothed_points)
```

```python
import matplotlib.pyplot as plt

# Define the input points and smoothing thresholds
points = [(0, 0), (1, 0), (2, 0), (3, 0), (3, 1), (3, 2), (3, 3), (2, 3), (1, 3), (0, 3), (0, 2), (0, 1)]
angle_threshold = 5 # degrees
distance_threshold = 0.2 # units

# Apply point boundary smoothing to the input points
smoothed_points = point_boundary_smoothing(points, angle_threshold, distance_threshold)

# Plot the original points and smoothed points on a graph
x, y = zip(*points)
plt.plot(x, y, '-o', label='Original Points')
x, y = zip(*smoothed_points)
plt.plot(x, y, '-o', label='Smoothed Points')
plt.legend()
plt.show()
```
